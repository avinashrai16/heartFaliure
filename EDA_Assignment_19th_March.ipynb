```python
#Q1 What is the distribution of age among heart failure patients in the dataset?

plt.figure(figsize=(15,5))
sns.countplot(x = df['age'], data=df, palette='viridis',hue=df['age'],legend=False)
plt.title(f"distribution of age")
# Add labels and title
plt.xlabel('Age')
plt.ylabel('count')
plt.xticks(rotation=90)

# Show the plot
plt.show()

# insight :
# as per the given data set maximim number of heart patients belongs to 40 to 70 year range.

```

```python
#Q2. How does the death rate vary with age?

# indicates whether the patient died during the follow-up period (binary: 0 for no, 1 for yes).

df.groupby('age')['DEATH_EVENT'].mean().plot(kind='bar',figsize=(15,4))

# Add labels and title
plt.xlabel('Age Group')
plt.ylabel('Death Rate')
plt.title('Death Rate by Age Group')

# Show the plot
plt.show()

# insight :
# as per the given data set death rate is higher in old age patient above 79
```

```python
# Q3. What is the percentage of male and female patients in the dataset?

# Sex: Gender of the patient (binary: 0 for female, 1 for male).
temp = df.sex.value_counts(normalize=True)
print(f"percentage of Male patients in the dataset:{round(temp[1], 2)}")
print(f"percentage of Female patients in the dataset:{round(temp[0], 2)}")

```

```python
temp = pd.DataFrame(df.sex.value_counts(normalize=True))
# Plotting
plt.figure(figsize=(8, 6))
sns.barplot(x=temp.index, y=temp.sex, hue=temp.index)
plt.title('Percentage of Male and Female Patients')
plt.xlabel('Gender Category')
plt.ylabel('Percentage(%)')
plt.xticks(rotation=0)
# # Customize legend labels
plt.legend(title='Sex')
plt.show()

#insight:
# As per the data set there is more number of male(1) patients tham female(0) having heart failure
```

```python
#Q4. How does the platelet count vary among different age groups ?

df.groupby('age')['platelets'].mean().plot(kind='bar',figsize=(15,6))
# Add labels and title
plt.xlabel('Age Group')
plt.ylabel('platelets')
plt.title('platelets by Age Group')

# Show the plot
plt.show()
```

```python
# Q5 Is there a correlation between creatinine and sodium levels in the blood ?

df_cor = df[['creatinine_phosphokinase','serum_sodium']].corr()
df_cor
f,ax = plt.subplots(figsize=(10,4))
sns.heatmap(df_cor,annot=True, fmt=".2f", linewidth=0.5,linecolor="black")
plt.title("Corelation between creatinine_phosphokinase and serum_sodium")
plt.show()

# insight : creatinine_phosphokinase and serum_sodium is following a patternt as per the given dataset in the patients
```

```python
#Q6 how does the prevalence of high blood pressure differ between male and female patients?
#
df_copy = df.copy(deep=True)
df_copy['hue_category'] = pd.cut(df_copy['sex'], bins=2)
plt.figure(figsize=(10,5))
sns.countplot(x = df_copy['high_blood_pressure'], data=df_copy, palette='viridis',hue=df_copy['sex'])
plt.title(f"High blood pressure in Male and Female" )
# Add labels and title
plt.xlabel('High blood pressure')
plt.ylabel('count')

# Customize legend labels
legend_labels = ['Female', 'Male']  # Labels for each category
plt.legend(title='Sex', labels=legend_labels)
# Show the plot
plt.show()

#insight: Females patients are having less boold presssure compare to men as per the data set
```

```python
# Q7. what is the relationship between smoking habits and the occurrence of heart failure?
df_cor = df[df['smoking']==0]['DEATH_EVENT'].value_counts(normalize=True)
print(f"non somkers normalize data:")
print(df_cor)
print()
df_cor = df[df['smoking']==1]['DEATH_EVENT'].value_counts(normalize=True)
print(f"somkers normalize data:")
print(df_cor)
print()

df_copy = df.copy(deep=True)
plt.figure(figsize=(10,5))
sns.countplot(x = df_copy['smoking'], data=df_copy, palette='viridis',hue=df_copy['DEATH_EVENT'])
plt.title("Death Event in Smokers and Non Smokers")
# Add labels and title
plt.xlabel('smoking[0- Non Smokers, 1- Smokers]')
plt.ylabel('count')

# Show the plot
plt.show()

# # insight :
# as per the given data set there are hingh number of non smokers.
# approx 50% of the patient in both the smoking category died due heart failure , which tells that there is no direct realtion between smoking and heart failure
```

```python
#Q8: Are there any noticeable patterns in the distribution of death events across different age groups?

# filtering data where DEATH_EVENT is 1
df_filter = df[df['DEATH_EVENT']==1]

df_filter.groupby('age')['DEATH_EVENT'].count().plot(kind='bar',figsize=(15,6))
# Add labels and title
plt.xlabel('Age Group')
plt.ylabel('DEATH_EVENT')
plt.title('DEATH_EVENT by Age Group')

# Show the plot
plt.show()

# insight :
# heart failure death events are more in old age patients.
```

```python
#Q9 Is there any significant difference in ejection fraction between patients with and without diabetes ?
df_copy = df.copy(deep=True)
df_copy['hue_category'] = pd.cut(df_copy['diabetes'], bins=2)
plt.figure(figsize=(15,5))
sns.countplot(x = df_copy['ejection_fraction'], data=df_copy, palette='viridis',hue=df_copy['diabetes'])
plt.title("ejection fraction in patients with and without diabetes")
# Add labels and title
plt.xlabel('ejection fractions values across patients')
plt.ylabel('ejection_fraction count')

# Customize legend labels
legend_labels = ['non diabetic', 'diabetic']  # Labels for each category
plt.legend(title='diabetes', labels=legend_labels)
# Show the plot
plt.show()


# insight:
# ejection fraction levels are higher in non diabetic patients comapre to diabetic such as 65 and above
```

```python
# Q10 How does the serum creatinine level vary between patients who survived and those who did not?

df_copy = df.copy(deep=True)
df_copy['hue_category'] = pd.cut(df_copy['DEATH_EVENT'], bins=2)
plt.figure(figsize=(15,5))
sns.countplot(x = df_copy['serum_creatinine'], data=df_copy, palette='viridis',hue=df_copy['DEATH_EVENT'])
plt.title(f"serum creatinine in patients who survived and those who did not" )
# Add labels and title
plt.xlabel('serum creatinine levels in patients')
plt.ylabel('serum creatinine count')

# Customize legend labels
legend_labels = ['survived', 'died']  # Labels for each category
plt.legend(title='DEATH_EVENT', labels=legend_labels)
# Show the plot
plt.show()

# insight
#patients who survived are low level of serum creatinine compare to thoese who did not survive.
```
